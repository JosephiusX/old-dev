```python
today = dt.date(2022,5,31)

jan_2022 = dt.date(2022, 1, 1)
jan_2021 = dt.date(2021, 1, 1)

start = jan_2021
end = today

start, end
```

```python
aapl = pdr.DataReader('AAPL', 'yahoo',start=start, end=end)
amc = pdr.DataReader('AMC', 'yahoo',start=start, end=end)
amzn = pdr.DataReader('AMZN', 'yahoo',start=start, end=end)
crsp = pdr.DataReader('CRSP', 'yahoo',start=start, end=end)
docu= pdr.DataReader('DOCU', 'yahoo',start=start, end=end)
dg = pdr.DataReader('DG', 'yahoo',start=start, end=end)
gush = pdr.DataReader('GUSH', 'yahoo',start=start, end=end)
goog = pdr.DataReader('GOOG', 'yahoo',start=start, end=end)
shop = pdr.DataReader('SHOP', 'yahoo',start=start, end=end)
saia = pdr.DataReader('SAIA', 'yahoo',start=start, end=end)
cabo = pdr.DataReader('CABO', 'yahoo',start=start, end=end)
nflx = pdr.DataReader('NFLX', 'yahoo',start=start, end=end)
fb = pdr.DataReader('FB', 'yahoo',start=start, end=end)
ostk = pdr.DataReader('OSTK', 'yahoo',start=start, end=end)
fvrr = pdr.DataReader('FVRR', 'yahoo',start=start, end=end)
tsla = pdr.DataReader('TSLA', 'yahoo',start=start, end=end)

dia = pdr.DataReader('DIA', 'yahoo',start=start, end=end)
spy = pdr.DataReader('SPY', 'yahoo',start=start, end=end)
dji = pdr.DataReader('^DJI', 'stooq')

#  = pdr.DataReader('', 'yahoo',start=start, end=end)

# ixic = pdr.DataReader('IXIC', 'yahoo', start=start, end=end)
# btcusd = pdr.DataReader('BTCUSD', 'stooq')

def calc_cum_perc(df):
    df['Percent Change'] = 100*(df['Adj Close']-df['Adj Close'].iloc[0])/df['Adj Close'].iloc[0]
    return df

aapl = calc_cum_perc(aapl)
amc = calc_cum_perc(amc)
amzn = calc_cum_perc(amzn)
crsp = calc_cum_perc(crsp)
docu= calc_cum_perc(docu)
dg = calc_cum_perc(dg)
gush = calc_cum_perc(gush)
goog = calc_cum_perc(goog)
shop = calc_cum_perc(shop)
saia = calc_cum_perc(saia)
cabo = calc_cum_perc(cabo)
nflx = calc_cum_perc(nflx)
fb = calc_cum_perc(fb)
ostk = calc_cum_perc(ostk)
fvrr = calc_cum_perc(fvrr)
tsla = calc_cum_perc(tsla)

plt.figure(figsize=(10,3),dpi=200)


aapl['Percent Change'].plot(label='AAPL')
# amc['Percent Change'].plot(label='AMC')
amzn['Percent Change'].plot(label='AMZN') 
crsp['Percent Change'].plot(label='CRSP') 
docu['Percent Change'].plot(label='DOCU')
dg['Percent Change'].plot(label='DG') 
gush['Percent Change'].plot(label='GUSH') 
goog['Percent Change'].plot(label='GOOG') 
shop['Percent Change'].plot(label='SHOP') 
saia['Percent Change'].plot(label='SAIA') 
cabo['Percent Change'].plot(label='CABO') 
nflx['Percent Change'].plot(label='NFLX') 
fb['Percent Change'].plot(label='FB') 
ostk['Percent Change'].plot(label='OSTK') 
fvrr['Percent Change'].plot(label='FVRR') 
tsla['Percent Change'].plot(label='TSLA') 

plt.legend()

# all from jan 2021 till may 2022
```
